# **For Loop**
- It is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set or a string).
- Using Break & Continue keyword.


- for loop - for x in list:
- break	Loop - if x == 'stop': break
- continue - if x == 'skip': continue

```python
animals = ['tiger', 'elephant', 'deer', 'zebra', 'lion']

for x in animals:
    # print(x)
    print('I saw a', x)
```

```python
for x in animals:
  print(x)
  if x == 'deer':
    break
```

```python
for x in animals:
  if x == 'zebra':
    continue
  print(x)
```

# **While Loop**
- With the while Loop we can execute a set of statements as long as a condition is true.
- Using Break & Continue keyboard

```python
num = 1

while num < 6:
  print(num)
  num += 1
```

```python
num = 1

while num < 11:
  print(num)
  if num == 7:
    break
  num += 1
```

```python
num = 0

while num < 11:
  num += 1
  if num == 7:
    continue
  print(num)
```

# **Python Functions**
- A function is a block of code which only runs when it is called.
- Advanced Functions - lambda, map,  zip

```python
def my_function():                         # Functining Defining
  print("Hello I'm from a function")
```

```python
# def add(a, b, c):
#   return a+b+c

# add(4,5,1)

add = lambda a, b, c: a+b+c               # Using lambda function
add(4,5,1)
```

```python
nums = [1, 2, 3, 4, 5]         # Map Function

def sq(n):
  return n*n

square = map(sq, nums)
print(list(square))
```

```python
name = ['Preet', 'Nikhil', 'Disha']           # Zip Function
ID = [4, 1, 3]

mapped = zip(name, ID)
print(list(mapped))
```

**Bonus Challenge:**


- Har salary pe 10% increment apply karna hai (map use karke).

- Ek function likhna hai jo salary ke basis par category bataye:

 Above 50,000 → "High Salary"

 Between 30,000–50,000 → "Medium Salary"

 Below 30,000 → "Low Salary"

- Loop se har employee ka naam + updated salary + category print karna hai.


```python
Employees = ['Shivam', 'Roshan', 'Arjun', 'Omkar']
Salaries = [28000, 45000, 56000, 31000]

Updated_Salaries = list(map(lambda x: int(x * 1.1), Salaries))


def Categorize(salary):
  if salary > 50000:
    return "High Salary"
  elif salary >= 30000:
    return "Medium Salary"
  else:
    return "Low Salary"


for i in range(len(Employees)):
  print(Employees[i], "-", Updated_Salaries[i], "-", Categorize(Updated_Salaries[i]))

```
