# **Groupby Function**
.sum() → total

.mean() → average

.count() → number of entries

.max() / .min() → highest / lowest value

.agg() → multiple functions ek saath

```python
employee = pd.read_csv('/content/Employee data.csv')
```

```python
employee
```

```python
import pandas as pd

grouped = employee.groupby('Department')['Salary'].mean()
grouped
```

```python
grouped = employee.groupby('Department')['Salary'].agg(["mean", "sum", "max", "min"])
grouped
```

```python
grouped = employee.groupby(["Department", "Employee"])["Salary"].sum()

grouped
```

```python
grouped = employee.groupby("Department")

data = grouped.get_group("Technical")
data
```

# **Top Commands**
Creating New, Renaming & Droping Columns.

Apply Method, Sort Index, Sort Values.

```python
import pandas as pd

employee = pd.read_csv('/content/Employee data.csv')
employee
```

**Creating New Columns**

```python
employee['Working Months'] = [3, 4, 6, 2, 6, 4, 5, 3, 7, 8, 4, 6]
employee.head()
```

```python
employee['Total'] = employee['Salary'] * employee['Working Months']
employee.head()
```

**Renaming Columns**

```python
employee.columns = ['EMPLOYEE', 'DEPARTMENT', 'SALARY', 'EMPLOYMENT', 'WORKING MONTHS', 'TOTAL']
employee.head()
```

```python
employee.rename(columns = {'EMPLOYEE': 'NAME', 'TOTAL': 'TOTAL(amount)'}, inplace = True)
# inplace=True se original DataFrame update hota hai. (Permanently Delete)

employee.head()
```

**Dropping Columns**

```python
employee.drop(columns = ['EMPLOYMENT'], inplace = True)
employee.head()
```

**Apply Method (Custom Logic)**

```python
employee['GST'] = employee['SALARY'].apply(lambda x: x * 0.02)
employee.head()
```

**Sort Index**

index use for (date, id , roll no.)

```python
employee.index = ['EMP001', 'EMP002', 'EMP003', 'EMP004', 'EMP005', 'EMP006', 'EMP007', 'EMP008', 'EMP009', 'EMP010', 'EMP011', 'EMP012']

employee.sort_index().head()     # ascending = True

# employee.sort_index(inplace = True) Permanently

# employee.sort_index(ascending = False, inplace = True)
```

**Sort Values**

```python
employee.sort_values(by = 'SALARY', ascending = False).head()
```

**BONUS CHALLENGE**

```python
import pandas as pd

data = {
    "Name": ["Neha", "Radha", "Krishna"],
    "Salary": [30000, 35000, 40000],
    "Bonus": [2000, 2500, 3000]
}

df = pd.DataFrame(data)

# 1. Create new column "TotalPay"
df["TotalPay"] = df["Salary"] + df["Bonus"]

# 2. Rename "Bonus" to "AnnualBonus"
df.rename(columns={"Bonus": "AnnualBonus"}, inplace=True)

# 3. Drop "Salary" column
df.drop(columns=["Salary"], inplace=True)

# 4. Apply 10% tax deduction on "TotalPay"
df["NetPay"] = df["TotalPay"].apply(lambda x: x * 0.9)

# 5. Sort by "NetPay"
df.sort_values(by="NetPay", ascending=False, inplace=True)

print(df)
```
