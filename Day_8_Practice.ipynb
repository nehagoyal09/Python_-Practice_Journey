# **Pandas Series**
- A Pandas Series is like a column in a table.

- It is a one-dimensional array holding data of any type.

```python
import pandas as pd


s = pd.Series([1,3,7,8,9])                # Series method
print(s)
```

```python
s = pd.Series([1,3,7,8,9], index=['a','b','c','d','e'])    # series with index method
print(s)
```

```python
print(s['b'])
print(s['e'])
```

# **Pandas DataFrame**
A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns.

**Series** = ek column

**DataFrame** = table of multiple columns (collection of Series)




```python
data = {"a": 100, "b": 200, "c": 300, "d":400}
df = pd.Series(data)
print(df)
```

```python
import pandas as pd       # DataFrame using dictionary

data = {"Name": ['Alex', 'Harry', 'Peter', 'John'], "Age": [23, 26, 24, 27], "City": ['Mumbai', 'Delhi', 'Chennai', 'Bangalore']}
df = pd.DataFrame(data)
print(df)
```

```python
df.head()
```

```python
# DataFrame with index

data = {"Items": ['Pencil', 'Pen', 'Copy', 'Book'], "Price": [10, 20, 30, 40]}
df = pd.DataFrame(data, index=['a1','b2','c3','d4'])
print(df)
```

```python
df.head(2)
```

```python
df['Items']      # Column Selection
```

```python
print(df.loc['b2','Price'])    # Row Selection
print(df.iloc[3,0])
```

# **Pandas Concatenation**
It is used to concatenate pandas objects along a particular axis with optional set logic along the other axes.


**Series Concatenation** = ek ke baad ek values jodna

**DataFrame Concatenation** = rows ya columns ko merge karna

```python
import pandas as pd

df1 = pd.DataFrame({"A": [1,2], "B": [3,4]})
df2 = pd.DataFrame({"A": [5,6], "B": [7,8]})

result = pd.concat([df1,df2])
print(result)
```

```python
result = pd.concat([df1,df2], axis=1)
print(result)
```

```python
result = pd.concat([df1,df2], ignore_index=True)
print(result)
```

# **Pandas Indexing**



- **loc[]**	[Labels (row/column names)]   	df.loc[0, "Name"]
- **iloc[]**	[Integer positions]	df.iloc[0, 0]
- **Filtering**	 [Boolean conditions]	df[df["Age"] > 23]

```python
data = {"Name": ['Alex', 'Harry', 'Peter', 'John'], "Age": [23, 26, 24, 27], "City": ['Mumbai', 'Delhi', 'Chennai', 'Bangalore']}
df = pd.DataFrame(data)

print(df.loc[0])                  # loc method
```

```python
print(df.loc[0, "Name"])
```

```python
print(df.loc[[0,2], ["Name","City"]])
```

```python
print(df.iloc[0])                      # iloc method
```

```python
print(df.iloc[0,0])
```

```python
print(df.iloc[[0,2], [0,2]])
```

```python
print(df[df["Age"] > 23])                 # Filter Method
```

```python
print(df[df["City"] == "Delhi"])
```

```python
print(df[(df["Age"] > 23) & (df["City"] == "Mumbai")])

```

```python
print(df[(df["Age"] > 23) | (df["City"] == "Delhi")])
```
