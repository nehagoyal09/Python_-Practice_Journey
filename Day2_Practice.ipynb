# **List and it's Methods**
- Lists are used to store multiple items in a single variable.
- It is a collection which is ordered, changeable and allows duplicate members.

**List Methods**

append(), sort(), reverse(), remove(), pop(), insert(), index(), extend(), count(), copy(), clear()

```python
fruits = ['apple', 'cherry', 'banana']

fruits.append("orange")                         # add item on last position
fruits.sort()                                   # 'asc to des' & 'desc to asc'
fruits.reverse()                                # arrange wise reverse
fruits.remove("banana")
fruits.pop(1)                                   # delete 0th, 1th element
fruits.insert(1, "watermelon")                  # add 0th, 1th element
fruits.index("orange")                          # show position

others = ['pineapple', 'strawberry', 'guava', 'cherry', 'pinepple']

fruits.extend(others)
fruits.count("cherry")
fruits.copy()
fruits.clear()
```

# **Conditional Statements**
- if, elif, else statements

```python
a,b = 330, 200

if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
else:
    print("a is greater than b")
```

# **Bonus Challenge (Mix all three)**

```python
students = ['Harry', 'John', 'Alex']
marks = [65, 82, 97]

for i in range(len(students)):
    name = students[i]
    score = marks[i]

    if score >= 90:
        remark = "Excellent"
    elif score >= 75:
        remark = "Good"
    else:
        remark = "Needs Improvement"

    print(name.upper(), "→", remark)
```
