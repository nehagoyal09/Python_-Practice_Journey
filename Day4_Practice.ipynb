# **Tuple**
- They are used to store multiple items in a single variable.
- A tuple is a collection which is ordered and unchangeable. It is written with round brackets.

```python
fruits = ('apple', 'cherry', 'papaya', 'banana')       # Standard Method
print(fruits)
```

```python
fruits = ('apple', 'cherry', 'papaya', 'apple')        # Allows Duplicates
print(fruits)
```

```python
marks = tuple((69, 84, 72, 90))                       # Function Method
```

```python
marks
```

```python
nums = (1, 2, 3, 4, 1, 7, 5, 2, 1)             # Count Method

nums.count(2)
# nums.count(1)
```

```python
nums = (1, 2, 3, 4, 1, 7, 5, 2, 1)             # Index Method

nums.index(7)
```

```python
nums = (1, 2, 3, 4, 1, 7, 5, 2, 1)

# len(nums)                                       # Length Method
sorted(nums)                                      # Sorting Method
```

```python
nums = (1, 2, 3, 4, 1, 7, 5, 2, 1)

max(nums)                                 # Maximum Method
# min(nums)                                 # Minimum Method
```

# **Dictionary**

- Dictionaries are used to store data values in key:value pairs.
- A dictionary is a collection which is ordered*, changeable and do not allow duplictaes.

```python
car = {"brand": "Ford", "model": "Mustang", "year": 1964}

car.keys()                # keys Method

# car.values()              # Values Method

# car.get('model')          # Get Method

# car.items()               # Items Method
```

```python
# car.pop("model")          # Pop Method

# car.popitem()             # Popitem Method
```

```python
car = {"brand": "Ford", "model": "Mustang", "year": 1964}

car.update({"color": "Black"})        # Update Method
print(car)

# car.clear()                           # Clear Method
# print(car)
```

# **Set**
- Sets are used to store multiple items in a single variable.
- Sets are written with curly brackets.

```python
fruits = {'apple', 'cherry', 'papaya', 'banana'}

fruits.add('mango')                       # Add Method
print(fruits)

# fruits.remove('cherry')                     # Remove & Discard Method
# print(fruits)

# fruits.pop()                               # Pop Method
# print(fruits)
```

```python
fruits = {'apple', 'cherry', 'papaya'}
company = {'google', 'microsoft', 'apple'}

print(fruits.union(company))                  # combine both sets together to form new set

# print(fruits.intersection(company))           # print common element from both sets

# print(fruits.difference(company))               # output will be from LHS of dot set only

# print(company.difference(fruits))
```
